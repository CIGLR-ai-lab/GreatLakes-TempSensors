```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
# Create a figure and add a subplot with a projection
fig, ax = plt.subplots(figsize=(10, 10),
                       subplot_kw={'projection': ccrs.LambertConformal(central_longitude=-86.0)})

# Set extent to cover the Great Lakes region
ax.set_extent([-93, -76, 40, 50])

# Add and shade land
# NOTE: The order in which you add these features matters.
# For example, if you move this command after the lakes one, you'll get a blank map
land_feature = cfeature.NaturalEarthFeature('physical', 'land', '10m', edgecolor='face',
                                            facecolor=cfeature.COLORS['land'])
ax.add_feature(land_feature)

# Add coastlines
ax.add_feature(cfeature.COASTLINE)

# Add Great Lakes major lakes feature
great_lakes = cfeature.NaturalEarthFeature(category='physical', name='lakes', scale='10m')
ax.add_feature(great_lakes, edgecolor=None)

# Define coordinates for the point you want to plot (I used the example point you selected)
point_longitude = -87.793 # Longitude of the point
point_latitude = 48.061   # Latitude of the point

# Plot the point as a plus('+')
ax.plot(point_longitude, point_latitude, '+', color='white', markersize=10, transform=ccrs.Geodetic())

# Add a gridlines and labels to the plot
gl = ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
gl.top_labels = False
gl.right_labels = False

plt.show()
```
